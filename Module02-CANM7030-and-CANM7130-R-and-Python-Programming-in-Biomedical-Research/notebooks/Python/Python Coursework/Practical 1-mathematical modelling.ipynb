(e) The measured mean can be used as an estimator µ¯ for the expected value. But how good is this
estimate? Take the entirety of (a) (i.e. drawing Ndraw = 10 rv’s and measuring the mean) as a single
experiment. Perform this experiment Nexp = 100 times and measure the σµ¯ standard deviation of these
results. This is an estimate of the standard deviation of the measured mean. What happens if we
increase Ndraw to 100, 200, 500? What happens if we increase Nexp? Can you figure out the dependence
of σµ¯ on Ndraw?

```python
import random
import math

def one_experiment_mean(Ndraw):
    return sum(random.randint(50, 100) for _ in range(Ndraw)) / Ndraw
one_experiment_mean(10)

```

```python

def sigma_mean(Ndraw, Nexp=100):
    means = [one_experiment_mean(Ndraw) for _ in range(Nexp)]
    m = sum(means)/Nexp
    var = sum((x - m)**2 for x in means) / (Nexp - 1)
    return math.sqrt(var)

for Ndraw in [10, 100, 200, 500]:
    print(Ndraw, sigma_mean(Ndraw, Nexp=100))
```

As more experiments are ran, the standard deviation around the mean decreases and the means becomes more stable.

Problem 2. Next let us look at drawing rv’s from some common probability distributions.
(a) Draw 500 random variables from a Gaussian distribution with mean µ = 2000 and variance σ
2 = 103.
Display a histogram of the resulting rv's.

```python

import numpy as np
import matplotlib.pyplot as plt

# parameters
mu = 2000
sigma = np.sqrt(1000)
N = 500

# empty list to store random variables
rv = []

# loop to generate Gaussian random variables
for i in range(N):
    x = np.random.normal(mu, sigma)
    rv.append(x)

# plot histogram
plt.hist(rv, bins=30, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 500 Gaussian Random Variables (μ = 2000, σ² = 1000)")
plt.show()

```

(b) Do the same for an exponential distribution with rate λ= 0.5. Measure the mean and the standard
deviation of the resulting distribution. What do you expect to obtain from these measurements?

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
lam = 0.5
N = 500

rv = []

# generate exponential random variables
for i in range(N):
    x = np.random.exponential(scale=1/lam)
    rv.append(x)

rv = np.array(rv)

plt.hist(rv, bins=30, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 500 Exponential Random Variables (λ = 0.5)")
plt.show()

```
