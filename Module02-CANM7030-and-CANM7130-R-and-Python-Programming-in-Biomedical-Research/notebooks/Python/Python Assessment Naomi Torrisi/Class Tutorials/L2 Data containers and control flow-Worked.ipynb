# 4

```python
# Set a gene dictionary as requested
geDi = {"TP53" : "ACGAGGGCCGAGTGGCGCAG",
       "THBS1" : "GGTGGCGCAGACGAGGGCCG",
       "MTOR" : "CCCAGGGCCGAGTGGCGCAG",
       "GAPDH" : "TTTAGGGGGGGAGTGGCGCAG",
       "NOTCH1" : "ACGAGGGCCGAGTGGCGCAG",
       "SF3B1" : "AAAAAGGGCCGAGTGGCGCAG",
       "DLL4" : "ACGAGGTTTTAGTGGCGCAG",
       "NFIL3" : "ACGAGCCCCGTGGCGCAG",
       "MOS" : "ACGAGGGCGGGGGGGCGCAG",
       "OCT4" : "TTTTGGGCCGAGTGGCGTTTT"}

# Set the new empty dictionary
newgeDi = {}

# Start a for loop for the dictionary keys
for i in geDi:
    # Set an empty dictionary to hold the counts
    freq = {}
    # Get the sequence for the ith key
    opera = geDi[i]
    
    
    # Use a for loop to count the bases
    for bp in set(opera):
        # Add the count to frequency dictionary
        freq[bp] = opera.count(bp)
        #print(freq) # This is commented out
    newgeDi[i] = freq
# Display the whole thing
newgeDi
```
