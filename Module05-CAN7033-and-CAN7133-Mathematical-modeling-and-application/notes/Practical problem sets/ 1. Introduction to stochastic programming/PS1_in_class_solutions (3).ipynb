```python
def randExp(n):
    indList = []
    for k in range(n):
        randNum = rd.randint(50,100)
        indList.append(randNum)
    meanOfList = sum(indList) / len(indList)
    return meanOfList

def mean(list):
    return sum(list)/len(list)
    

def standardDev(list):
    var = 0
    for el in list:
        var += (el - mean(list))**2
    var = var/len(list)
    return var**0.5


nExps = 100
meanList = []
for i in range(nExps):
    meanCur = randExp(40)
    meanList.append(meanCur)
print(meanList)

standardDev(meanList)
```
