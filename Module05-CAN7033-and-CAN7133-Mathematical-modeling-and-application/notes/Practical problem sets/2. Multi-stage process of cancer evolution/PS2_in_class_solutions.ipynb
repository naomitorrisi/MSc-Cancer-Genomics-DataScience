```python
import random as rd 

gaussVars = []
for i in range(500):
    var = rd.gauss(2000, 103**0.5)
    gaussVars.append(var)

gaussVars = [rd.gauss(2000, 103**0.5) for i in range(500)]

print(gaussVars)

import matplotlib.pyplot as plt
plt.hist(gaussVars, 20)
plt.xlabel("x")
plt.ylabel("number of counts")
plt.show()
```

```python

def countUntilFail(p):
    success = False
    numberOfFails = 0
    while not success:
        success = rng.random()<p
        numberOfFails += 1

    return numberOfFails+1

countUntilFail(0.01)

def waitingTime(p, dt):
    return countUntilFail(p)*dt

def expectedWaitingTime(p, dt):
    waitingTimesList = [waitingTime(p, dt) for i in range(1000)]
    return np.mean(waitingTimesList)

expectedWaitingTime(0.1, 1)

_p = np.arange(0.1,0.8,0.05)
print(_p)

expectedWT_p = [expectedWaitingTime(_p[i], 1) for i in range(len(_p))]

expectedWT_p = []
for p in _p:
    expectedWT_p.append(expectedWaitingTime(p, 1))

print(expectedWT_p)

plt.plot(_p, expectedWT_p)
plt.xlabel("p")
plt.ylabel("ev_p")
plt.show()




```
