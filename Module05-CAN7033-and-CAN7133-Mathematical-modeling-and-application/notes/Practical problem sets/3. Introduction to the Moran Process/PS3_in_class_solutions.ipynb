```python
import numpy as np
rng = np.random.default_rng()

N = 1
P = 0.2
T = 10
dt = 0.3
t = 0
N_t = [N]
_t = [t]
while t<T:
    t+=dt
    # --- for every cell, check whether it: ---
    for i in range(N):
        # --- divides ---
        if rng.random()<P:
            N+=1

    # --- remember current values of N and t for plotting ---
    N_t.append(N)
    _t.append(t)

import matplotlib.pyplot as plt

plt.plot(_t,N_t)
plt.show()
```

```python
import numpy as np
rng = np.random.default_rng()

def runExperiment():
    N = 1
    P = 0.2
    T = 10
    dt = 0.3
    t = 0
    N_t = [N]
    _t = [t]
    while t<T:
        t+=dt
        for i in range(N):
            if rng.random()<P:
                N+=1
        N_t.append(N)
        _t.append(t)
    return _t, N_t

_t, N_t = runExperiment()
_t, N2_t = runExperiment()
_t, N3_t = runExperiment()

gamma = 0.2 / dt
f_t = []
for t in _t:
    value = np.exp(gamma*t)
    f_t.append(value)

print(f_t)

plt.plot(_t, f_t)
plt.plot(_t, N_t)
plt.plot(_t, N2_t)
plt.plot(_t, N3_t)
plt.xlabel("time")
plt.ylabel("number of cells")
plt.show()

```
