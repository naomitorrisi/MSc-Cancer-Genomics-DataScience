```python
import numpy as np
rng = np.random.default_rng()


def experiment(lam, T):
    # lam = 0.5
    # T = 10
    t = 0
    count = 0
    while t<T:
        waitingTime = rng.exponential(1/lam)
        t = t + waitingTime
        # print(t)
        count += 1
        # do something
    divisions = count - 1
    return divisions


lam = 0.5
T = 10

nDivisions = []
for i in range(5000):
    result = experiment(lam, T)
    nDivisions.append(result)

nPoisson = []
for i in range(5000):
    result = rng.poisson(lam*T)
    nPoisson.append(result)

import matplotlib.pyplot as plt
plt.hist(nDivisions, 50)
plt.hist(nPoisson, 50)
xlabel="number of divisions"
ylabel="counts"
plt.show()
```

```python
# ---- Moran model ----



def probMutantBirth(X, N):
    return X/N

def probMutantDeath(X, N):
    return X/N

def selectBirth(P):
    if rng.random()<P:
        return "mutant"
    else:
        return "wildtype"

def selectDeath(P):
    if rng.random()<P:
        return "mutant"
    else:
        return "wildtype"

T = 100
N = 10
X = 5
Y = N-X
X_t = [X, ]
for i in range(T):
    # select mutant or wild type for birth
    Pbirth= probMutantBirth(X,N)
    Pdeath = probMutantDeath(X,N)
    if selectBirth(Pbirth)=="mutant":
        X += 1
    
    if selectDeath(Pdeath)=="mutant":
        X -= 1

    X_t.append(X)

print(X_t)

plt.plot(range(len(X_t)), X_t)
plt.show()
```
